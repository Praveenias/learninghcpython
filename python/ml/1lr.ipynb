```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_excel('prices.xlsx')
df.area
```

```python
%matplotlib inline
plt.xlabel("price")
plt.ylabel("area")
plt.scatter(df.price,df.area,marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.predict([[3000]])
```

```python
reg.coef_
```

```python
reg.intercept_
df.area
```

```python
%matplotlib inline
plt.xlabel("area")
plt.ylabel("price")
plt.scatter(df.area,df.price,marker='+')
plt.plot(df.area,reg.predict(df[['area']]),color='blue')
```
