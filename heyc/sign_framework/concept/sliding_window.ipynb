```python
#max sum of array k subarray
a =  [2, 1, 5, 1, 3, 2]
k = 3
start = 0
max_sum = 0
current_sum = 0

for end in range(len(a)):
  current_sum += a[end]
  if end-start+1 == k:
    max_sum = max(max_sum,current_sum)
    current_sum -= a[start]
    start +=1
print(max_sum)
```

```python
# Longest Subarray with Sum â‰¤ 
arr = [2, 1, 5, 2, 8]
S = 7
start = 0
max_length=0
current_sum = 0
for end in range(len(arr)):
  current_sum += arr[start]
  while current_sum >S:
    current_sum -= arr[start]
    start+=1
  max_length = max(max_length, end - start + 1)
  print(arr[start:end+1],max_length)
```

```python
arr = [2, 1, 5, 2, 3, 2]
target = 7
start = 0
mins = float('inf')
for end in range(len(arr)):
  temp = arr[start:end+1]
  while sum(temp) >=target:
    print(temp)
    mins = min(mins,end-start+1)
    start+=1
    temp = arr[start:end+1]
print(mins)
```
