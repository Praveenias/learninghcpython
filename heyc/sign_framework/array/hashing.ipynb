```python
#Number of good pairs
def numIdenticalPairs(nums) -> int:
  hash = {}
  count = 0
  for i in nums:
    if i in hash:
      count += hash[i]
      hash[i] += 1
    else:
      hash[i] = 1
    
  return count
  

#print(numIdenticalPairs([1,2,3,1,1,3]) == 4)
print(numIdenticalPairs([1,1,1,1]))
# numIdenticalPairs([1,2,3])
```
