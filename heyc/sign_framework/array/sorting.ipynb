```python
#How Many Numbers Are Smaller Than the Current Number
def smallerNumbersThanCurrent(nums):
  sortedNums = sorted(nums)
  l1 = []
  for i in range(len(nums)):
    num = nums[i]
    l1.append(sortedNums.index(num))
  return l1

print(smallerNumbersThanCurrent([8,1,2,2,3]))
print(smallerNumbersThanCurrent([6,5,4,8]))
print(smallerNumbersThanCurrent([7,7,7,7]))
```

```python
#2037. Minimum Number of Moves to Seat Everyone

def minMovesToSeat(seats,students):
  sorted_seats = sorted(seats)
  sorted_students = sorted(students)
  count = 0
  print(sorted_seats,sorted_students)
  for i,j in zip(sorted_seats,sorted_students):
    if i == j:
      continue
    if j > i:
      count += j-i
    else :
      count += i-j

  return count


# print(minMovesToSeat([3,1,5],[2,7,4]) ==4)
# print(minMovesToSeat([4,1,5,9],[1,3,2,6])==7)
# print(minMovesToSeat([2,2,6,6],[1,3,2,6])==4)
print(minMovesToSeat([12,14,19,19,12],[19,2,17,20,7])==19)
```

```python


def bubble_sort(items):
  for i in range(len(items)):
    already_sorted = True
    for j in range(len(items)-i-1):
      if items[j] > items[j+1]:
        items[j],items[j+1] = items[j+1],items[j]
        already_sorted = False
    if already_sorted:
      break
  return items

print(bubble_sort([3,2,4,5,1,3]))
```

```python
def insertion_sort(items):
  for i in range(1,len(items)):
    j = i-1
    c_item = items[j]
    while (j>=0 and c_item < items[j]):
      items[j+1] = items[j]
  print(items)

print(insertion_sort([8,2,6,4,5]))
```
