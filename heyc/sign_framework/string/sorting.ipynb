```python
#1. Sorting the sentence

def sortSentence(s: str) -> str:
  s = s.split()
  out = ['']*len(s)
  for i in s:
    pos = int(i[-1])
    out[pos-1] = i[:-1]
  print(' '.join(out))

print(sortSentence("is2 sentence4 This1 a3"))
```

```python
def frequencySort(s: str) -> str:
  freq = {}
  for i in s:
    if i in freq:
      freq[i] += 1
    else:
      freq[i] = 1
 
  sorted_ar = sorted(freq.items(),key=lambda x:x[1],reverse=True)
  
  res = ''
  for s1 in sorted_ar:
    res += s1[0]*s1[1]
  return res



frequencySort("aAbb")
```

```python
def min_difference(scores, k):
    if k == 1:
        return 0  # Only one score, no difference.
    
    scores.sort()  # Step 1: Sort the scores
    n = len(scores)
    min_diff = float('inf')  # Initialize minimum difference as infinity
    print(scores)
    
    # Step 2: Use sliding window to find minimum difference
    for i in range(n - k + 1):
        diff = scores[i + k - 1] - scores[i]
        min_diff = min(min_diff, diff)
    
    return min_diff

# Example Usage
scores = [90, 85, 75, 60, 120]
k = 3
print(min_difference(scores, 2)) 
```
