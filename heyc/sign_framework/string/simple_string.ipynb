```python
#1. Final value of variable after performing operations
operations = ["X++","++X","--X","X--"]
x = 0
for i in operations:
  if i in ["--X","X--"]:
    x -=1
  if i in ["++X","X++"]:
    x +=1
print(x)
```

```python
#3. Maximum number of words found in sentences

l1 = ["alice and bob love leetcode","i think so too","this is great thanks very much"]
#out = max([len(i.split()) for i in l1])
#print(out)
max = 0
for s1 in l1:
    print(len(s1.split()))
    count = len(s1.split())
    print(count)
    if count > max:
        max = count
```

```python
#4. Goal parser interpretation

command = "G()(al)"
replaceD = {
  "G":"G",
  "()":"o",
  "(al)":'al'
}
#return command.replace("()","o").replace("(al)","al")
temp = ''
res = ''
for i in range(len(command)):
  temp += command[i]
  if temp in replaceD:
    res += replaceD[temp]
    temp=''
print(res)
```
