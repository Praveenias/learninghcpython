```python
#1. Jewels and stones
def numJewelsInStones(self, jewels: str, stones: str) -> int:
        count = 0
        for stone in stones:
            if stone in jewels:
                count +=1
        return count 
        
```

```python
#2. Count the number of consistent strings

def countConsistentStrings(allowed, words) -> int:
        count = 0
        allowed = set(allowed)
        for word in words:
            if not set(word).difference(allowed):
                count +=1
        return count

print(countConsistentStrings("ab",["ad","bd","aaab","baa","badab"]) == 2)
print(countConsistentStrings("abc",["a","b","c","ab","ac","bc","abc"]) == 7)
```

```python
#3 Rings and rods

def countPoints(rings) -> int:
  holder = {i:set() for i in range(10)}
  count = 0
  for j in range(0,len(rings),2):
    ring = int(rings[j+1])
    if len(holder[ring]) == 3:
      continue
    holder[ring].add(rings[j])
    if len(holder[ring]) == 3:
      count +=1
  return count
    

print(countPoints("B0B6G0R6R0R6G9") == 1)
print(countPoints("B0R0G0R9R0B0G0") == 1)
print(countPoints("G4") == 0)
```

```python
#4. Check if the sentence is pangram

def checkIfPangram(sentence: str) -> bool:
  dict = {i:False for i in range(97,123)}
  count = 0
  for i in sentence:
    dict[ord(i)] = True
  return all(dict.values())


print(checkIfPangram("thequickbrownfoxjumpsoverthelazydog")==True)
print(checkIfPangram("leetcode")==False)
```
